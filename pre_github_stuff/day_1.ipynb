```python
def process_line(line):
    line = line.strip()
    value = int(line[1:])
    
    if line[0] == '+':
        return(value)
    elif line[0] == '-':
        return(-value)
    else:
        raise

def chronal_calibration(input_file_path, starting_value=0):
    with open(input_file_path) as file:
        input_text = file.readlines()
        
    frequency = starting_value
    for line in input_text:
        #print("current freq:", frequency)
        #print("input line:",line)
        value = process_line(line)
        #print("value:",value)
        frequency += value
        #print('_______')
    
    return(frequency)


```

```python
def chronal_calibration_for_duplicates(unprocessed_list, starting_value=0):
    processed_input = []
    for line in unprocessed_list:
        processed_line = process_line(line)
        processed_input.append(processed_line)
    
    adjustment_cycle = itertools.cycle(processed_input)
    
    frequencies_set = set()
    old_length = len(frequencies_set)
    
    current_frequency = starting_value
    frequencies_set.add(current_frequency)
    new_length = len(frequencies_set)
    #print(old_length, new_length)
    
    while old_length < new_length:
        #print(old_length, new_length)
        #print(current_frequency)
        #print(frequencies_set)
        old_length = new_length
        current_frequency += next(adjustment_cycle)
        frequencies_set.add(current_frequency)
        new_length = len(frequencies_set)
        #print ('___')
    
    return(current_frequency)
```
